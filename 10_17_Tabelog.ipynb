```python
# step0: import
import pandas as pd
import urllib.request as req
import bs4 as bs

# step1:
datas = []
for i in range(5):
  page = i + 1
  print("頁數:", page)
  url = "https://tabelog.com/tw/tokyo/rstLst/" + str(page) + "/?SrtT=rt"
  response = req.urlopen(url)
  html = bs.BeautifulSoup(response)

  rs = html.find_all("div", {"class":"list-rst__body"})
  for r in rs:
    en = r.find("a", {"class":"list-rst__rst-name-target"})
    rating = r.find("span", {"class":"c-rating__val"})
    intro = r.find("p", {"class":"list-rst__pr-title"})
    prices = r.find_all("span", {"class":"c-rating-v3__val"})

    en_text = en.text
    en_href = en["href"]
    rating_text = rating.text
    if intro == None:
      intro_text = "無"
    else:
      intro_text = intro.text
    dinner_price = prices[0].text
    lunch_price = prices[1].text
    intro_text = intro_text.strip()

    print(rating_text)
    print(en_text)
    print(en_href)
    print(intro_text)
    print(dinner_price)
    print(lunch_price)
    print("-" * 30)

    # step2: 取得清單
    d = {
        "rating":rating_text,
        "en": en_text,
        "intro":intro_text
    }
    datas.append(d)

# step3: 存檔 list->DataFrame(pandas表格型態)
df = pd.DataFrame(datas)
df.to_csv("tabelog.csv",
      # sep="\t", #TSV/CSV
      index=False, #關閉列標籤
      columns=["rating", "en", "intro"]
)
```
