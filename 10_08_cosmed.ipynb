````python
import urllib.request as req
import time

while True:
  url = 'https://www.cosmed.com.tw/api/getStore.aspx?t=store&c=&d=&s='
  f = req.urlopen(url)
  s = f.read()
  s = s.decode()
  if 'Invalid' in s:
    print('retry...')
    # 間隔3秒無間斷重送
    time.sleep(3)
  else:
    break;

print(s)
````
